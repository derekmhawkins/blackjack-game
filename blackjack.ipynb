```python
from random import shuffle
from IPython.display import clear_output

class Deck:
    def __init__(self):
        self.suit = ["Spades", "Clubs", "Hearts", "Diamonds"]
        self.numbers = [i for i in range(1, 13)]
        self.sets = [(i, j) for i in self.suit for j in self.numbers]
        
    def get_decks(self, num):
        self.sets *= num
        return self.sets
    
    def shuffle_deck(self):
        shuffle(self.sets)
        return self.sets
        
        
class Player:
    def __init__(self):
        self.hand = []
        self.cards = 0
        self.tally = 0
    
    def get_total(self):
        self.tally = 0
        for i in self.hand:
            self.tally+=i[1]
        if self.tally == 21:
            return self.blackjack()
        return int(self.tally)
        
    def bust(self):
        print(f"{self.__str__()} Bust\n{'=' * 30}")
    
    def show_hand(self, cls):
        if isinstance(self, Human):
            print(f"{cls.__str__()} hand: {self.hand}: {self.get_total()}")
        elif isinstance(self, Dealer) and len(self.hand) < 3:
            print(f"{cls.__str__()} hand: {self.hand[1]}: {self.hand[1][1]}")
        elif isinstance(self, Dealer) and len(self.hand) > 2:
            print(f"{cls.__str__()} hand: {self.hand}: {self.get_total()}")
            
    def blackjack(self):
        return "Blackjack!"
    

class Human(Player):
    def __init__(self):
        super().__init__()
        
    def __str__(self):
        return f"Player"
    

class Dealer(Player):
    def __init__(self):
        super().__init__()
    
    @classmethod
    def deal_hand(self, cls, deck):
        while(cls.cards <= 1):
            cls.hand.append(deck.sets.pop())
            cls.cards+=1
        return cls.hand
    
    @classmethod
    def hit(self, cls, deck):
        cls.hand.append(deck.sets.pop())
        return cls.hand
    
    def end_hand(self):
        print(f"{self.__str__()} hand: {self.hand}: {self.get_total()}")
    
    def __str__(self):
        return f"Dealer"
    

class Game:
    def __init__(self):
        self.dealer = Dealer()
        self.human = Human()
        self.players = [self.dealer, self.human]
    
    @classmethod
    def win(self, cls):
        print(f"{cls} wins!\n{'=' * 30}")
    
    def push(self):
        print("Push")
        
#     def num_players(self, num):
#         self.players = [self.dealer, self.human * num]

def main():
    deck = Deck()
    player = Human()
    dealer = Dealer()
    game = Game()
    
    num_decks = int(input("How many decks do you want to play with? "))
    deck.get_decks(num_decks)
    print(f"You are playing with {num_decks} decks.")
    
    not_shuffled = True
    while(not_shuffled):
        if not_shuffled:
            print("Shuffling...")
            deck.shuffle_deck()
            break
        else:
            not_shuffled = False
    print("Deck is shuffled")
#     print(f"Here's your deck: \n{deck.sets}")
    print("Dealer dealing...")
    dealer.deal_hand(player, deck)
    player.show_hand(player)
#     print(f"Here's your deck: \n{deck.sets}")
    dealer.deal_hand(dealer, deck)
    dealer.show_hand(dealer)
    
    while True:
        ask = input("Would you like to hit, or stand? ")
        clear_output()
        if ask == 'hit'.lower():
            dealer.hit(player, deck)
            player.show_hand(player)
            dealer.show_hand(dealer)
        if player.get_total() > 21:
            clear_output()
            player.bust()
            game.win(dealer)
            break
        if player.get_total().__int__() == 21:
            player.blackjack()
            continue
        if ask == 'stand'.lower():
            dealer.hit(dealer, deck)
            player.show_hand(player)
            dealer.show_hand(dealer)
            
            less_than_16 = True
            while less_than_16:
                if dealer.get_total().__int__() <= 16:
                    clear_output()
                    dealer.hit(dealer, deck)
                    player.show_hand(player)
                    dealer.show_hand(dealer)
                    continue
                else:
                    less_than_16 = False
            if dealer.get_total().__int__() == player.get_total().__int__():
                clear_output()
                game.push()
                break
            if dealer.get_total().__int__() > 21:
                clear_output()
                dealer.bust()
                game.win(player)
                break

            if dealer.get_total().__int__() < player.get_total().__int__():
                clear_output()
                game.win(player)
                break
            else:
                clear_output()
                game.win(dealer)
                break

    player.show_hand(player)
    dealer.end_hand()
```

```python
main()
```
